# Strings

You've already seen plenty of strings in examples during the previous lessons, but just to recap, strings in Python can be defined using either single or double quotatons. They are functionally equivalent.

```python
x = 'Pluto is a planet'
y = "Pluto is a planet"

x == y
```

Double quotes are convenient if your string contains a single quote character (e.g. representing an apotrophe).

If we try to put a single quote character inside a single-quoted string, Python gets confused. We can fix this by **escaping** the single quote with a backslash.

```python
'Pluto\'s a planet!'
```

The table below summarizes some important uses of the backslash character.

| What you type... | What you get... | Example        | Print     
|------------------|-----------------|----------------|----------------|
| \\' | ' | 'What\\'s up?' | What's up?|
| \\" | " | "That's \\"cool\\"" | That's "cool"|
| \\\ | \ | "Look, a mountain: /\\\" | Look, a mountain: /\ |
| \\n |   | "1\\n2 3" | 1 2 3|

The las sequence, \n, represents the *newline character*. It causes Python to start a new line.

In addition, Python's triple quote syntax for strings lets us include newlines literally (i.e. by just hitting 'Enter' on our keyboard, rather than using the special '\n' sequence). We've already seen this in the docstrings we use to document our functions, but we can use them anywhere we want to define a string.

```python
hello = "hello\nworld"
triple_quoted_hello = """hello
world"""

print(triple_quoted_hello)

triple_quoted_hello == hello
```

Strings can be thought of as sequences of characters. Almost everything we've seen that we can do to a list, we can also do to a string.

```python
# Indexing
planet = 'Pluto'
planet[0]
```

```python
# Slicing
planet[-3:]
```

```python
# how long is this string?
len(planet)
```

```python
# Yes, we can even loop over them
[
    char + '!'
    for char in planet
]
```

A major way in which they differ from lists is that they are *immutable*. we can't modify them.
